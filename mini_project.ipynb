```python
#Mini Project- Random Password Generator!
from random import randint
password = ""
for i in range(10):
    i = chr(randint(65,98))
    password =str(password)+i
    print(password)
    
    password=""
    for i in range(5):
        i=chr(randint(65,98))
        for j in range(5):
            j=chr(randint(65,98)).lower()
            password=str(password)+i+j
            print(password)
```
